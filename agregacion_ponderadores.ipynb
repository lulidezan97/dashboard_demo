```python
import pandas as pd
import numpy as np
```

```python
df_p= pd.read_csv(r"campaña2_ponderadores.csv")
```

```python
base= pd.read_csv(r"version1407.csv")
```

```python
encuesta= base.drop(["circ_limpia","act_limpia","filepath","circulacion_abierta","submitter","act_original"], axis=1)
```

```python
pond= df_p[["edad_muestra","ponderador_masc","ponderador_fem"]]
```

```python
df_p
```

```python
pond.dropna(inplace=True)
```

```python
pond.drop(6,inplace=True)
```

```python
pond
```

```python
categorical_features = ["opinion_cuarentena","home_office","opinion_circulacion"]
```

```python
for col in categorical_features:
    dummies = pd.get_dummies(encuesta[col], prefix=col)
    encuesta = pd.concat([encuesta, dummies], axis=1)
    encuesta.drop(col, axis=1, inplace=True)
```

```python
encuesta
```

```python
agreg= encuesta.groupby(["edad","genero"]).sum().reset_index()
```

```python
agreg= agreg.loc[((agreg.genero == "Hombre") |(agreg.genero=="Mujer")) & (agreg.edad!= "Menor de 18 años")].reset_index(drop=True)
```

```python
agreg
```

```python
pond_ord= pd.melt(pond,id_vars=["edad_muestra"]).sort_values(by="edad_muestra")
```

```python
pa= agreg.iloc[:,2:].apply(lambda x: x*pond_ord["value"])
```

```python
pa["genero"] = agreg["genero"]
pa["edad"]= agreg["edad"]
```

```python
masc= pa.loc[pa.genero == "Hombre"].reset_index(drop= True)
```

```python
fem= pa.loc[pa.genero == "Mujer"].reset_index(drop= True)
```

```python
dfs= [fem, masc]
```

```python
for df in dfs:
    df.loc["subtotal"] = df.sum(axis=0, numeric_only= True)
```

```python
a= pa.groupby(["edad","genero"]).sum().reset_index()
```

```python
a= pd.melt(a,id_vars=["edad","genero"])
```

```python
a['pregunta'] = a.variable.apply(lambda x: ' '.join(x.split('_')[:2]))
```

```python
a['respuesta'] = a['variable'].str.split('_').str[2]
```

```python
a.drop("variable", axis=1, inplace=True)
```

```python
a['porcentaje']=np.nan
```

```python
dfs= [fem, masc]
for df in dfs:
    df.drop("total_grupo",axis=1, inplace=True)
    df.drop("subtotal",inplace=True)
```

```python
for i in set(a.pregunta):
    lista=a.loc[a.pregunta==i,'value']
    subtotal=sum(lista)
    a.loc[a.pregunta==i,'porcentaje']=a.loc[a.pregunta==i,'value']/subtotal
```

```python
a.to_csv(r"femasc_ponderado.csv",index=False)
```
